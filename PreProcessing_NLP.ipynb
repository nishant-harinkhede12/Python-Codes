In this step all the important libraries for lemmatization and stop words are imported

```python
nltk.download('wordnet')
from nltk.stem import wordnet
from nltk.stem import WordNetLemmatizer
```

```python
import re # re libray will use for regular expression

# ps = PorterStemmer()
wordnet=WordNetLemmatizer()
sentences = nltk.sent_tokenize(gold)
nltk.download('stopwords')
from nltk.corpus import stopwords
corpus = []
```

```python
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    review = review.lower()
    review = review.split()
#    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]   
    review = ' '.join(review)
    corpus.append(review)
```

In Below step Name entity recognization is performed

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
from nltk import ne_chunk
corpus = ne_chunk(corpus)
```

```python
c = pd.DataFrame(corpus)
c.head()
```

To create Bag of words

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
```
