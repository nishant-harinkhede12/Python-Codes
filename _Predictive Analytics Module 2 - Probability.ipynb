```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(4.5,3)})
from random import seed
from scipy.stats import uniform
```

```python
seed(1)
n = 10000
a = 0
b = 10
data_inform = uniform.rvs(size=n, loc=a, scale =b)
```

```python
ax = sns.distplot(data_inform,bins = 100)
ax.set(xlabel = ('Uniform'),ylabel = ('frequency'))
plt.show()
```

```python
from scipy.stats import expon
data_expon = expon.rvs(scale =1, loc=0, size=1000)

ax = sns.distplot(data_expon, bins = 100)

ax.set(xlabel = ('Uniform distribution'),ylabel = ('frequency'))
```

```python
from scipy.stats import norm
data_norm = norm.rvs(size=10000,loc=0,scale=1)

ax = sns.distplot(data_norm, bins = 100)
```
