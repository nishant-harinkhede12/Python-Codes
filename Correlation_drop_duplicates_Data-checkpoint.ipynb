```python
import pandas as pd
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

chipo = pd.read_csv(url, sep = '\t')
```

```python
chipo.info()
```

```python
prices = [float(value[1: -1]) for value in chipo.item_price]
chipo.item_price = prices
chipo.info()
```

```python
chipo_filter = chipo.drop_duplicates(['item_price','quantity'])
print(len(chipo))
print(len(chipo_filter))
```

```python
chipo_1_prod = chipo_filter[chipo_filter['quantity'] == 1]
print(len(chipo_1_prod))
```
