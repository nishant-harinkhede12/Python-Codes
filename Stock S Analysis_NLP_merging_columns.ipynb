```python
import pandas as pd

url = 'https://raw.githubusercontent.com/krishnaik06/Stock-Sentiment-Analysis/master/Data.csv'
data = pd.read_csv(url,encoding = 'unicode_escape')
data.head(5)
```

IDENTIFYING THE MIN/MAX VALUES FROM DATE COLUMN TO DIVIDE DATA INTO TRAIN/TEST

```python
print(max(data['Date']))
print(min(data['Date']))
```

```python
train = data[data['Date'] < '20150101']
test = data[data['Date'] > '20141231']
```

```python
# train.head()
```

# Removing punctuations

```python

train_data=train.iloc[:,2:27]
train_data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

```

```python
train_data.head()
```

# Renaming column names for ease of access

```python
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
train_data.columns= new_Index
train_data.head(5)
```

# Convertng headlines to lower case


```python
for index in new_Index:
    train_data[index]=train_data[index].str.lower()
train_data.head(5)
```

SAMPLE CODE TO JOIN MULTIPLE COLUMN FOR SPECIFIC ROW

```python
' '.join(str(x) for x in train_data.iloc[1,0:25])
```

THIS STEP WILL JOIN ALL THE COMMENTS COLUMN TO MAKE ONE HEADLINE

```python
headlines = []
for row in range(0,len(train_data.index)):
    headlines.append(' '.join(str(x) for x in train_data.iloc[row,0:25]))
```

```python
# headlines[0::5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

## IMPLEMENTING BAG OF WORDS

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
 train.isnull().sum()
# train['Label'].value_counts()
```

```python
train['Label'].fillna(train['Label'].mode()[0],inplace=True)
```

```python
test['Label'].fillna(test['Label'].mode()[0],inplace=True)
```

# implement RandomForest Classifier

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

## Predict for the Test Dataset


```python
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
